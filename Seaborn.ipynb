```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# some lists
y = [4, 6, 8, 9]
x = ['A', 'B', 'C', 'D']
```

```python
# plt.xticks(rotation=30);
# plt.legend('str','str')
plt.figure(figsize=(5, 3))
plt.title('Диаграмма столбчатая (bar)')
plt.xlabel('Category')
plt.ylabel('Quantity')
sns.barplot(x = x, y = y, width=0.3);
```
