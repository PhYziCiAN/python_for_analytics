```

    new_sum = 0
    for key, value in kwargs.items():
        new_sum += value
    print(new_sum)
total_fruits(banana=5, orange=6,mango=8,apple=7,cocnut=3)
```

````python
def total_fruits(**kwargs):
    new_sum = 0
    for key, value in kwargs.items():
        new_sum += value
    print(new_sum)
````

#### Задание2. Создание словаря из 2 списков

````python
keys = ['One', 'Two', 'Three']
values = [1, 2, 3]
````

````python
%%time

new_dict = {keys[i]:values[i] for i in range (len(keys))}
print(new_dict)
````
