```python
df['delta_renovated'] = df['yr_renovated']-df['yr_built']
df.loc[df['delta_renovated']<0,'delta_renovated'] = 0
df.delta_renovated.sort_values()
```

```python
df['date'] = df['date'].str.replace('T','')
df['yr_sale'] = pd.to_datetime(df['date'], format='%Y%m%d%H%M%S').dt.year
df['month_sale'] = pd.to_datetime(df['date'], format='%Y%m%d%H%M%S').dt.month
d = dict(enumerate(calendar.month_abbr))
df['month_sale'] = df['month_sale'].map(d)
df.head(20)
```
